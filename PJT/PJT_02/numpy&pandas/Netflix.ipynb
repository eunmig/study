```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv_path = "data/NFLX.csv"

df = pd.read_csv(csv_path, usecols=range(0, 5))

df
```

```python
df.dtypes
```

```python
# 날짜 데이터 변환
df["Date"] = pd.to_datetime(df["Date"])
# 2021년 이후 데이터 필터링
new_df = pd.read_csv(csv_path, usecols=range(0, 5))
df_after_2021 = new_df[new_df["Date"] >= "2021-01-01"]
df_after_2021["Date"] = pd.to_datetime(df_after_2021["Date"])

# 그래프 그리기 (가로, 세로 축에 표시될 데이터를 차례로 기입)
plt.plot(df_after_2021['Date'], df_after_2021['Close'])


# 그래프 제목 설정
plt.title('NFLX Close Price')

# x축 레이블 설정
plt.xlabel('Date')

# x축 회전
plt.xticks(rotation=45)

# y축 레이블 설정
plt.ylabel('Close')

# 그래프 표시
plt.show()
```

```python
# 날짜 데이터 변환
df["Date"] = pd.to_datetime(df["Date"])

# 2021년 이후 데이터 필터링
new_df = pd.read_csv(csv_path, usecols=range(0, 5))
df_after_2021 = new_df[new_df["Date"] >= "2021-01-01"]
df_after_2021["Date"] = pd.to_datetime(df_after_2021["Date"])

# 월 별로 그룹화
Average_Close = df_after_2021.groupby(pd.Grouper(key = "Date", freq = 'M'))["Close"].mean().reset_index()

# 그래프 그리기 (가로, 세로 축에 표시될 데이터를 차례로 기입)
plt.plot(Average_Close['Date'], Average_Close['Close'])

# 그래프 제목 설정
plt.title('Monthly Average Close Price')

# x축 레이블 설정
plt.xlabel('Date')

# x축 회전
plt.xticks(rotation=45)

# y축 레이블 설정
plt.ylabel('Average Close Price')

# 그래프 표시
plt.show()
```

```python
# 날짜 데이터 변환
df["Date"] = pd.to_datetime(df["Date"])

# 2021년 이후 데이터 필터링
new_df = pd.read_csv(csv_path, usecols=range(0, 5))
df_after_2021 = new_df[new_df["Date"] >= "2022-01-01"]
df_after_2021["Date"] = pd.to_datetime(df_after_2021["Date"])

# 그래프 그리기
plt.plot(df_after_2021['Date'], df_after_2021['High'], label='High')
plt.plot(df_after_2021['Date'], df_after_2021['Low'], label='Low')
plt.plot(df_after_2021['Date'], df_after_2021['Close'], label='Close')

# 그래프 제목 설정
plt.title('High, Low,and Close Prices since january 2022')

# x축 레이블 설정
plt.xlabel('Date')

# x축 회전
plt.xticks(rotation=45)

# y축 레이블 설정
plt.ylabel('Price')

# 범례 표시
plt.legend()

# 그래프 표시
plt.show()
```
